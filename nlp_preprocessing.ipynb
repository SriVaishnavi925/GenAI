```python
import re
from collections import Counter
```

```python
def preprocess_text(text):
    if not text or text.strip() == "":
        return [], ""

    # Remove URLs & emails
    text = re.sub(r"http\S+|www\S+|@\S+", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Lowercase
    text = text.lower()

    # Handle repeated characters (soooo → so)
    text = re.sub(r'(.)\1+', r'\1', text)

    # Remove special characters & emojis
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()

    # Remove short words (keep no, not)
    tokens = [w for w in tokens if len(w) > 2 or w in ["no", "not"]]

    clean_sentence = " ".join(tokens)

    return tokens, clean_sentence
```

1. "Love" vs "love"
They are treated as different tokens due to case sensitivity. Lowercasing is used to normalize them.

2. If stopwords are not removed
They increase noise, data size, and reduce model efficiency.

3. When removing stopwords is harmful

Sentiment analysis: "not good" becomes "good"
Question answering: removing words affects meaning and context

4. Stemming vs Lemmatization

Stemming: cuts words roughly (running → run)
Lemmatization: uses dictionary, more accurate (better → good)

```python
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product ",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for s in sentences:
    tokens, clean = preprocess_text(s)
    results.append((s, tokens, clean))

    print("Original:", s)
    print("Tokens:", tokens)
    print("Clean:", clean)
    print("-"*40)
```

```python
for s, tokens, clean in results:
    if tokens:
        avg_len = sum(len(t) for t in tokens) / len(tokens)
    else:
        avg_len = 0

    print("Sentence:", s)
    print("Total Tokens:", len(tokens))
    print("Unique Tokens:", len(set(tokens)))
    print("Avg Token Length:", avg_len)
    print("-"*40)
```

```python
all_tokens = []

for _, tokens, _ in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 frequent:", counter.most_common(10))
print("Least 5 frequent:", counter.most_common()[-5:])
```

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```
